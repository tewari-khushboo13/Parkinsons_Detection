```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv('train_data.csv')
X = df.drop('1.1', axis=1)
y = df['1.1']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel
lsvc = LinearSVC(C=0.5, penalty="l1", dual=False, random_state=42).fit(X_train, y_train)
model = SelectFromModel(lsvc, prefit=True)
X_train_new = model.transform(X_train)
X_test_new = model.transform(X_test)

```

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel
lsvc = LinearSVC(C=0.5, penalty="l2", dual=False, random_state=42).fit(X_train, y_train)
model = SelectFromModel(lsvc, prefit=True)
X_train_new = model.transform(X_train)
X_test_new = model.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
model = Sequential()
model.add(Dense(units=40, activation='relu', input_dim=X_train_new.shape[1]))
model.add(Dense(units=30, activation='relu'))
model.add(Dense(units=40, activation='relu'))
model.add(Dense(units=30, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train_new, y_train, epochs=100, batch_size=10, verbose=1)
loss, accuracy = model.evaluate(X_test_new, y_test)
print(f'Test Accuracy: {accuracy*100:.2f}%')
```
